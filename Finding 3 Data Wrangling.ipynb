```python
import pandas as pd
import numpy as np
```

```python
worldcup = pd.read_csv('WorldCups.csv')
```

```python
worldcup["Host Country"] = worldcup["Year"].map(str) + ' ' + worldcup["Country"]
```

```python
worldcup['Winner']=np.where(worldcup['Winner'] == 'Germany FR','Germany', worldcup['Winner'])
worldcup['Runners-Up']=np.where(worldcup['Runners-Up'] == 'Germany FR','Germany', worldcup['Runners-Up'])
worldcup['Third']=np.where(worldcup['Third'] == 'Germany FR','Germany', worldcup['Third'])
worldcup['Fourth']=np.where(worldcup['Fourth'] == 'Germany FR','Germany', worldcup['Fourth'])
```

```python
worldcup['rank'] = np.where(worldcup['Country'] == worldcup['Winner'],1\
                            ,np.where(worldcup['Country'] == worldcup['Runners-Up'],2\
                            ,np.where(worldcup['Country'] == worldcup['Third'],3\
                            ,np.where(worldcup['Country'] == worldcup['Fourth'],4,0))))
```

```python
worldrank = pd.read_csv('World Rank.csv')
missingrank = pd.read_csv('Missing Ranks.csv')
```

```python
df = worldcup
df1 = worldrank
df2 = missingrank
```

```python
df2.index = df.loc[df['rank']==0].index
```

```python
df.loc[df['rank'] == 0, 'rank'] = df2['rank']
```

```python
df = df.merge(df1,on='Year')
```

```python
df['Different'] = df['world rank'] - df['rank']
```

```python
df['Difficulty'] = df['Different'] / df['rank']
```

```python
df = df[['Year','Host Country','Winner','Runners-Up','Third','Fourth','rank'\
               ,'world rank','Different','Difficulty']]
```
